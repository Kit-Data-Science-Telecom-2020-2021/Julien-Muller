```python
# import des modules usuels
import numpy as np
import pandas as pd
from math import *
import json 

# options d'affichage
pd.set_option("display.min_rows", 16)
```

```python
# chargement des données
geo = pd.read_csv("correspondance-code-insee-code-postal.csv", sep=';', usecols=range(11))
geo
```

```python
# Ajout d'une colonne 'CP Ville' avec le Code postal + un espace + et le nom de la Ville.
geo.insert(0,'CP Ville',geo['Code Postal'] + ' ' + geo['Commune'])
geo
```

```python
# méthode extract
geo[['Latitude', 'Longitude']] = geo["geo_point_2d"].str.extract("(.*), (.*)").astype(float)
geo.head()
```

Si l’on considère deux points A et B sur la sphère, de latitudes $\phi_{A}$ et $\phi_{B}$ et de longitudes $\lambda_{A}$ et $\lambda_{B}$ alors la distance angulaire $S_{A-B}$ en radians entre A et B est donnée par : 

$
\begin{equation}
S_{A-B} = arcos(sin(\phi_{A})sin(\phi_{B}) + cos(\phi_{A})cos(\phi_{B})cos(\lambda_{B} - \lambda_{A}))
\end{equation}
$

La distance S en mètres, s’obtient en multipliant $S_{A-B}$ par un rayon de la Terre conventionnel (6 378 137 mètres par exemple).

```python
# Ecrivez une fonction qui détermine la commune la plus proche d'un point à partir de sa latitude et sa longitude.

#Fonction renvoyant la distance en kilomètres entre 2 points A(lat1,long1) et B(lat2, long2) 
def distance_points_sphere(lat1,long1,lat2,long2):
    d_lambda = long2 - long1
    return 6378137*acos(sin(lat1)*sin(lat2)+cos(lat1)*cos(lat2)*cos(d_lambda))/1000

# Fonction renvoyant les informations correspondant à la commune la plus proche du point donné
def ville(lat, long):
    dico_distances = {Commune_name:distance_points_sphere(lat,long,lat_commune,long_commune)
                     for Commune_name,lat_commune,long_commune 
                      in zip(geo['Commune'],geo['Latitude'],geo['Longitude'])}
    
    Closer_commune, Distance_min = min(dico_distances.items(),key=lambda x: x[1])
    print("La commune la plus proche est " + str(Closer_commune) + " située à " + str(Distance_min) + " km du point considéré")
    return geo.loc[geo["Commune"] == Closer_commune].T
    
```

```python
# on applique la fonction à une coordonnée tirée au hasard
np.random.seed(0)
a, b = 41.5, 51.1  # latitude min et max de la France métropolitaine
lat = (b - a) * np.random.random() + a
a, b = -5.1, 9.5  # longitude min et max de la France métropolitaine
long = (b - a) * np.random.random() + a

ville(lat, long)
```

Afin d'utiliser la première fonction avec un GPS, nous devons construire un outil de conversion des coordonnées (Degrés, minutes, secondes) vers (latitude, longitude)

```python
# conversion degrés, minutes, secondes => décimal
def dms2dec(deg, mn, sec):
    return deg + mn/60 + sec/3600
```

```python
# à partir de coordonnées GPS précises
ville(dms2dec(48, 42, 52), dms2dec(2, 14, 45))
```

La colonne "geo_shape" comporte des chaines de catactères au format JSON. Elles représentent les formes géométriques des communes qui sont soit des polygones soit composées de plusieurs polygones.

- Utiliser la librairie Python json pour parser la colonne "geo_shape".
- Donner le décompte des valeurs accédées avec la clé "type".
- Donner le décompte des longueurs des listes accédées avec la clé "coordinates".
- Quelle commune est la plus complexe géométriquement ?
- Quelle commune est la seconde la plus complexe géométriquement ?
- Quelles sont les villes qui sont de type "Polygon" mais dont la longueur des listes accédées avec la clé "coordinates" vaut 2 ?
- Pour ces villes vérifier que le premier polygone contient bien le second (enclave). NB : installer la librairie shapely, utiliser la classe Polygon de shapely.geometry. Sur Windows shapely peut nécessiter d'installer la dll "geos_c.dll" dans le répertoire "Library/bin" de votre environnement Python.

```python
# Utiliser la librairie Python json pour parser la colonne "geo_shape".
# Ajout d'une colonne correspondant à chaque clé des json
geo["Type of shape"] = geo["geo_shape"].apply(lambda x: json.loads(x)["type"])
geo["Coordinates"] = geo["geo_shape"].apply(lambda x: json.loads(x)["coordinates"])
geo["Number of coordinates"] = geo["geo_shape"].apply(lambda x: len(json.loads(x)["coordinates"]))
geo
```

```python
# Donner le décompte des valeurs accédées avec la clé "type".
geo["Type of shape"].value_counts()
```

```python
# Donner le décompte des longueurs des listes accédées avec la clé "coordinates".
geo["Number of coordinates"].value_counts()
```

```python
# Quelle commune est la plus complexe géométriquement ?
geo.loc[geo["Number of coordinates"] == max(geo["Number of coordinates"].values),'Commune']
```

```python
# Quelle commune est la seconde la plus complexe géométriquement ?
# Ici j'utilise la conversion de geo["Number of coordinates"].values en set pour avoir toutes les valeurs distinctes
# De plus le set est déjà trié dans l'ordre croissant
# Enfin je reconvertis en liste afin d'obtenir la 2ème valeur en partant de la fin
geo.loc[geo["Number of coordinates"] == list(set(geo["Number of coordinates"].values))[-2],'Commune']
```

```python
# Quelles sont les villes qui sont de type "Polygon" mais dont la longueur des listes accédées 
#avec la clé "coordinates" vaut 2 ?
geo.loc[(geo["Type of shape"] == "Polygon") & (geo["Number of coordinates"]==2),'Commune']
```

```python
#Pour ces villes vérifier que le premier polygone contient bien le second (enclave). 
#NB : installer la librairie shapely, utiliser la classe Polygon de shapely.geometry. 
#Sur Windows shapely peut nécessiter d'installer la dll "geos_c.dll" 
#dans le répertoire "Library/bin" de votre environnement Python.

from shapely.geometry import Polygon
series_of_communes = geo.loc[(geo["Type of shape"] == "Polygon") & (geo["Number of coordinates"]==2),'Commune']
for commune in series_of_communes:
    Coordinates_record = geo.loc[geo['Commune'] == commune,'Coordinates']
    polygon1 = Polygon(Coordinates_record.values[0][0])
    polygon2 = Polygon(Coordinates_record.values[0][1])
    print(commune + '---> ' + str(polygon2.within(polygon1)))   
```
